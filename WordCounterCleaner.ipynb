```python
import collections
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize, word_tokenize
import csv
import pandas as pd
```

```python
# file open
data = open('DiscoCompleto.txt').read()

# change 'english' to 'spanish' when necessary 
stopWords = set(stopwords.words('spanish'))
words = word_tokenize(data)
wordsFiltered = []

for w in words:
    if w not in stopWords:
        wordsFiltered.append(w)

# Filter out punctuation and common words left behind like 're or 've
punctuation = {'.',',','\"','"','¿','?','!','-','—','(',')',"'", ':',";",'...',"jt","'d","'re","ta","'ve","'s","'t","ai","'m","n't","'ll","na","ca","wo"}
finalFilter = []
for word in wordsFiltered:
    if word not in punctuation:
        finalFilter.append(word)


# uncomment to check
#print(finalFilter)
```
